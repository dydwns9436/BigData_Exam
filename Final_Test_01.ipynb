Q1.  Tm 컬럼은 각 팀의 이름을 의미한다. TOR팀의 평균나이를 소수 4째 자리까지 구하여라

```python
a = df.groupby('Tm').get_group('TOR')
```

```python
b = df[df.Tm == 'TOR'].Age.mean()
print(round(b, 4))
```

```python
print(round(a['Age'].mean(),4))
```

Q2. Pos 컬럼은 포지션을 의미한다. 전체 선수 중 최소나이대의 선수들을 필터하고 

그들 중 가장 많은 포지션은 무엇인지 확인하라

```python
q2 = df[df.Age == df.Age.min()].Pos.value_counts().index[0]
print(q2)
```

Q3. 선수들의 이름은 first_name+ 공백 + last_name으로 이루어져 있다. 

가장 많은 first_name은 무엇이며 몇 회 발생하는지 확인하라

```python
df['Player'].str.split(' ').str[0].value_counts().index[0]
```

Q4.  PTS컬럼은 경기당 평균득점수 이다. 각포지션별로 경기당 평균득점수의 평균을 구하여라

```python
q4 = df.groupby('Pos')['PTS'].mean()
print(q4)
```

Q5. G컬럼은 참여한 경기의 숫자이다. 

각 (팀별)로 (가장 높은 경기참여 수)를 가진 (선수들의 경기 참여 숫자의 평균)을 구하여라

```python
q5 = df.groupby('Tm')['G'].max().mean()
print(q5)
```

Q6. Tm의 값이 MIA이며 Pos는 C또는 PF인 선수의 MP값의 평균은?

```python
q6 = df[(df.Tm == 'MIA') & (df.Pos.isin(['C', 'PF']))].MP.mean()
print(q6)
```

Q7. 전체 데이터중 G의 평균값의 1.5배 이상인 데이터들만 추출했을때 3P값의 평균은?

```python
df['G'].mean()
```

```python
q7 = df[df.G >= (df.G.mean()*1.5)]['3P'].mean()
print(q7)
```

Q8.  Age의 평균 이상인 그룹과 평균 미만인 그룹간의 G값의 평균의 차이는?

```python
a = df[df.Age >= df.Age.mean()].G.mean()
b = df[df.Age < df.Age.mean()].G.mean()

print(abs(a-b))
```

Q9. 평균나이가 가장 젊은 팀은 어디인가

```python
q9 = df.groupby('Tm').Age.mean().sort_values().index[0]
print(q9)
```

Q10. Pos그룹별 평균 MP값을 구하여라

```python
q10 = df.groupby('Pos')['MP'].mean()
print(q10)
```
