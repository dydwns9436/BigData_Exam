# 데이터 불러오기

# 문제

##Q1.  전체데이터의 수축기혈압(최고) - 이완기혈압(최저)의 평균을 구하여라

```python
q1 = (df['수축기혈압(최고) : mmHg'] - df['이완기혈압(최저) : mmHg']).mean()

print(q1)
```

##Q2. 50~59세의 신장평균을 구하여라

```python
q2 = df[(df.측정나이 >=50) & (df.측정나이 < 60)]['신장 : cm'].mean()

print(q2)
```

##Q3. 연령대 (20~29 : 20대 ...) 별  인원수를 구하여라


```python
df['연령대'] = df['측정나이']//10 * 10
q3 = df['연령대'].value_counts()

print(q3)
```

##Q4.  연령대 (20~29 : 20대 ...) 별 등급의 숫자를 데이터 프레임으로 표현하라

```python
q4 = df.groupby(['연령대','등급'], as_index = False).size()

print(q4)
```

##Q5. 남성 중 A등급과 D등급의 체지방률 평균의 차이(큰 값에서 작은 값의 차)를 구하여라

```python
df_M = df[df.측정회원성별 == 'M']
q5 = df_M.groupby('등급')['체지방율 : %'].mean()

print(abs(q5[0]-q5[-1]))
```

##Q6. 여성 중 A등급과 D등급의 체중의 평균의 차이(큰 값에서 작은 값의 차)를 구하여라

```python
df_F = df[df.측정회원성별 == 'F']
q6 = df_F.groupby('등급')['체중 : kg'].mean()

print(abs(q6[0]-q6[-1]))
```

##Q7. bmi는 자신의 몸무게(kg)를 신장/100의 제곱(m)으로 나눈값이다. 데이터의 bmi 를 구한 새로운 컬럼을 만들고 남성의 bmi 평균을 구하여라

```python
df['bmi'] = df['체중 : kg'] / (df['신장 : cm']/100) **2
q7 = df[df.측정회원성별 == 'M']['bmi'].mean()
print(q7)
```

##Q8.  bmi보다 체지방율이 높은 사람들의 체중평균을 구하여라

```python
q8 = df[df['체지방율 : %'] > df['bmi']]['체중 : kg'].mean()

print(q8)
```

##Q9. 남성과 여성의 악력 평균의 차이를 구하여라

```python
q9 = df_M['악력D : kg'].mean() - df_F['악력D : kg'].mean()

print(q9)
```

##Q10.남성과 여성의 교차윗몸일으키기 횟수의 평균의 차이를 구하여라

```python
q10 = df_M['교차윗몸일으키기 : 회'].mean() - df_F['교차윗몸일으키기 : 회'].mean()

print(q10)
```
