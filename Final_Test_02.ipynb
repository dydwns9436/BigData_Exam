Q1.  마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 가장 많은 인원을 가진 나이대는? (0-9 : 0 , 10-19 : 10)

```python
q1 = (df['age']//10).value_counts().index[0]
print(q1*10)
```

Q2. 마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 가장 많은 나이대 구간의 인원은 몇명인가?

```python
q2 = (df['age']//10).value_counts().max()
print(q2)
```

Q3. 나이가 25살 이상 29살 미만인 응답 고객들중 housing컬럼의 값이 yes인 고객의 수는?

```python
q3 = df[df.age.isin(range(25, 29)) & (df.housing =='yes')].shape[0]
print(q3)
```

Q4.  numeric한 값을 가지지 않은 컬럼들중 unique한 값을 가장 많이 가지는 컬럼은?

```python
lst = []
df_2 = df.select_dtypes('object')
for i in df_2:
    tar = df_2[i]
    lst.append([df_2, tar.nunique()])

q4 = pd.DataFrame(lst).sort_values(1, ascending = False).values[0][0]
print(q4)
```

```python
lst= [] 
for col in df.select_dtypes(exclude='int'):
    target = df[col]
    lst.append([col,target.nunique()])

result = pd.DataFrame(lst).sort_values(1,ascending=False).values[0][0]
print(result)
```

Q5. balance 컬럼값들의 평균값 이상을 가지는 데이터를 ID값을 기준으로 내림차순 정렬했을때 상위 100개 데이터의 balance값의 평균은?

```python
q5 = df[(df.balance >= df.balance.mean())].sort_values('ID', ascending = 0).head(100)['balance'].mean()
print(q5)
```

Q6. 가장 많은 광고를 집행했던 날짜는 언제인가? (데이터 그대로 일(숫자),달(영문)으로 표기)

```python
q6 = df[['day', 'month']].value_counts().index[0]
print(q6)
```

Q7. 데이터의 job이 unknown 상태인 고객들의 age 컬럼 값의 정규성을 검정하고자 한다. 샤피로 검정의 p-value값을 구하여라

```python
from scipy.stats import shapiro

q7 = shapiro(df[df.job == 'unknown']['age'])[1]
print(q7)
```

Q8.  age와 balance의 상관계수를 구하여라

```python
df[['age', 'balance']].corr().iloc[0,1]
```

Q9. y 변수와 education 변수는 독립인지 카이제곱검정을 통해 확인하려한다. p-value값을 출력하라

```python
# 넘 어렵다.. 카이제곱검정 함수는 처음 써본다..
from scipy.stats import chi2_contingency

v = pd.crosstab(df.y, df.education)
print(v)
chi2, p, dof, expected = chi2_contingency(v)
display(v)
print(p)
```

Q10. 각 job에 따라 divorced/married 인원의 비율을 확인 했을 때 그 값이 가장 높은 값은?

```python
t = df.groupby(['job', 'marital']).size().reset_index()
p_t = t.pivot_table(index='job', columns = 'marital')[0]
p_t.fillna(0, inplace= True)

p_t['ratio'] = p_t['divorced']/p_t['married']

print(p_t['ratio'].max())
```
