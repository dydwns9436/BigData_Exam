###1.해당 데이터셋에서 결측치가 가장 많은 변수를 찾아서 해당 결측치를 0으로 대치하고, 결측치를 제외한 평균과 0으로 대치한 후의 평균과의 차이를 구하시오.

```python
df = pd.read_csv('/content/drive/MyDrive/BigData/CSV_File/airquality.csv')
df.info()

#결측치를 0으로 대치 하기 전의 평균 구하기
mean = df['Ozone'].mean()

#결측치를 0으로 대치 한 후의 평균 구하기
df['Ozone'].fillna(0, inplace = True)
mean_0 = df['Ozone'].mean()

print(abs(mean_0 - mean))
```

###2.Wind 변수에 대해 최소 최대 정규화를 수행한 후 평균값과 Z 정규화를 수행한 후 평균 값의 차이를 구하시오.

```python
df = pd.read_csv('/content/drive/MyDrive/BigData/CSV_File/airquality.csv')

from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()

df['Wind_mms'] = mms.fit_transform(df[['Wind']])
a = df['Wind_mms'].mean()

df['Wind_z'] = (df['Wind'] - df['Wind'].mean()) / df['Wind'].std()
b = df['Wind_z'].mean()

print(abs(a-b))
```

### 데이터셋 : Bank_Personal_Loan_modeling.csv

Q. 은행에서 수집한 고객 5000명의 금융정보에 따른 대출여부가 들어있는 참조데이터를 이용해 대출 여부를 분류하는 가장 최적의 이웃의 크기값을 구하고, 이때 분류정확도를 산출하시오. 

또한 5000명의 사람들 중 은행에서 추가적인 관리가 필요한 특성을 가진 사람들의 수를 계산하시오. 


---


조건1 : 참조 데이터는 7:3의 비율로 훈련 데이터와 테스트 데이터를 구분

조건2 : 대출여부(y)는 계층적거으로 구분하는 샘플 구분 방식을 따름

조건3 :  Z값을 산출하여 정규화한다.

```python
df = pd.read_csv('/content/drive/MyDrive/BigData/CSV_File/Bank_Personal_Loan_Modelling.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df
```

```python
df = df.drop(['ID', 'ZIP Code'], axis=1)
```

```python
df = df.dropna()
```

```python
X = df.drop('Personal Loan', axis=1)
y = df['Personal Loan']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify = y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import Normalizer
nm = Normalizer()

X_train = nm.fit_transform(X_train)
X_test = nm.fit_transform(X_test)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

kn = KNeighborsClassifier()

kn.fit(X_train, y_train)

pred_kn = kn.predict(X_test)

acc = accuracy_score(y_test, pred_kn)
print(acc, kn.n_neighbors)
```
