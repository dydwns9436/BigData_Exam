##1.필요 패키지 임포트

- 넘파이 : import numpy as np

- 판다스 : import pandas as pd

- 맷플롯립 : import matplotlib.pyplot as plt

```python
import numpy as np        # 넘파이 패키지 임포트
import pandas as pd       # 판다스 패키지 임포트
import sklearn            # 싸이킷런 패키지 임포트

# 의사결정나무 분류모델을 위한 패키지 임포트
from sklearn.tree import DecisionTreeClassifier

# 학습 및 테스트 데이터셋 분리를 위한 패키지 임포트
from sklearn.model_selection import train_test_split
```

##2.데이터 불러오기

pd.read_csv('불러올 파일 디렉터리 또는 웹주소')

```python
# 깃허브에 있는 csv 파일을 읽어와서 데이터프레임 df로 넣는다.
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
```

##3.데이터 살펴보기

- df.head(), df.info(), df.shape, df.describe() 등을 사용해 데이터 구조를 살핌

```python
df
```

```python
df.shape
```

##4.데이터 전처리

```python
# 데이터 값을 변경학 위해 판다스 df의 replace() 함수를 이용
# 텍스트로 되어 있는 species 컬럼 데이터를 수치화

df['species'].replace({'setosa':0, 'versicolor':1, 'virginica':2}, inplace=True)
df
```

##5.분석 데이터셋 준비

- 전처리를 마친 데이터를 train data와 test데이터셋으로 분리

- 분리를 위해 사이킷런의 train_test_split() 함수를 이용

```python
# 분석 데이터셋 준비
# X는 독립변수(설명변수), y는 종속변수(목표변수)
X = df[["sepal_length","sepal_width","petal_length","petal_width"]]
y = df["species"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
#shape 속성을 이용해 결과 데이터셋 크기 확인
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

##6.데이터분석 수행

- 분석 문제는 크게 지도학습, 비지도학습으로 구분

- 지도학습은 분류와 예측 문제로 구분해서 모델 선택

> 지도학습, 비지도학습 분석기법

- 지도학습(분류) : 의사결정나무(분류), KNN, SVM, 로지스틱 회귀분석, 랜덤포레스트, 인공신경망

- 지도학습(회귀(예측)) : 선형회귀분석, 다중회귀분석, 의사결정나무(회귀)

- 비지도학습 : 군집분석, 연관분석, 인공신경망

```python
# 주어진 데이터로 붓꽃 조류를 구분하는 문제는 분류문제이며, 분류를 위한 알고리즘 중 의사결정나무를 이용
# DecisionTreeClassifier 객체를 생성하고 fit() 함수에 학습용 데이터 (X_train)와 결정값 데이터(y_train)를 입력해 호출하면 학습이 수행
dt = DecisionTreeClassifier(random_state=11)
dt.fit(X_train, y_train)    # 학습 수행
```

```python
# 학습이 완료된 dt 객체(모델)에서 테스트 데이터셋으로 분류(예측)를 수행
pred = dt.predict(X_test)
```

##7.성능평가 및 시각화

- 의사결정나무를 이용한 분류 모델의 성능 평가를 위한 대표적 지표는 정확도

- 사이킷런은 정확도 측정을 위해 accuracy_score() 함수를 제공

```python
# 모델 성능 - 정확도 측정
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc)
```

## 전체코드

```python
import numpy as np # 넘파이 패키지 임포트
import pandas as pd # 판다스 패키지 임포트
import sklearn # 사이킷런 패키지 임포트

from sklearn.tree import DecisionTreeClassifier # 의사결정나무 분류모델을 위한 패키지 임포트
from sklearn.model_selection import train_test_split # 학습 및 데이터셋 분리를 위한 패키지 임포트

df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')

df['species'].replace({'setosa':0, 'versicolor':1, 'virginica':2}, inplace = True) # species 컬럼의 데이터를 0, 1, 2로 변환

# 분석 데이터셋 준비
# X는 독립변수 y는 종속변수
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['species']

# 데이터셋 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)

# DecisionTreeClassifier 객체 생성
dt = DecisionTreeClassifier(random_state=11)

# 학습
dt.fit(X_train, y_train)

# 예측
pred = dt.predict(X_test)

# 모델 성능 - 정확도 측정
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc)
```

#3.분석모델 성능 평가 방법

##1.오차행렬

- 이진 분류의 예측 오류가 얼마인지와 어떠한 유형의 예측오류가 발생하는지를 함께 보여줌

- 사이킷런에서 confusion_matrix() 함수를 통해 구함

```python
# 의사결정나무를 이용한 붓꽃 분류 문제에서 분류모델의 오차행렬을 구하는 코드
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, pred)
```

- 가운데 대각선 숫자는 정확하게 분류한 숫자
- 30개의 테스트 데이터셋 중에 정확하게 분류한 데이터는 28개, 잘못 분류한 데이터는 2개

##2.모델 평가지표

- 사이킷런에서 오차행렬에 기반한 평가지표는 classification_report() 함수를 통해 구함

```python
# 모델 성능 평가 - 평가지표 계산
from sklearn.metrics import classification_report
rpt = classification_report(y_test, pred)
print(rpt)
```

> 정확도

- 정확도는 실제 데이터와 예측 데이터가 얼마나 정확한지를 판단하는 자료
- Accuracy : TP + TN / TP + TN + FP + FN

> 정밀도

- 정밀도는 Postive로 예측한 대상 중 실제값이 Postive로 일치하는 비율
- Precision : TP / TP + FP

> 재현율

- 재현율은 실제 Postive인 대상 중 실제와 예측 값이 일치하는 비율
- Recall : TP / TP + FN

> F1 Score

- 사이킷런의 f1_score 함수를 통해 구함
- 정밀도와 재현율을 결합한 조화평균 지표로 값이 클 수록 모형이 정확
- F1 : 2 * (정밀도 * 재현율 / 정밀도 + 재현율)
