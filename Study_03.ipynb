## 2.탐색적 데이터 분석(EDA)

### 2-1.EDA란?
- 수집한 데이터가 들어올 때 다양한 방법을 통해 자료를 관찰하고 이해하는 과정

### 2-2.필요성
- 데이터의 분포 및 값을 검토함으로써 데이터가 표현하는 현상을 이해하며 내재된 잠재적 문제를 인식하고 해결안을 도출

### 2-3.분석 과정 및 절차

1.분석 목적과 변수 확인

2.데이터의 문제성 확인

3.데이터의 개별 속성값 분포 확인

4.데이터 사이의 관계 확인

### 2-4.데이터 문제성 확인

1.결측치 대치 방법

- 단순대치법, 다중대치법 등

2.이상치 대치 방법
- 제거, 대체, 유지 등

### 2-5.데이터의 개별 속성값 분포 확인

*적절한 요약 통계지표 사용*

1.데이터의 중심
- 평균(mean), 중앙값(median), 최빈값(mode)

2.데이터의 분산
- 범위(range), 분산(var), 표준편차(std)

*사분위범위 방법 사용 등*
