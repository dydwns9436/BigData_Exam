## 2.탐색적 데이터 분석(EDA)

### 2-1.EDA란?
- 수집한 데이터가 들어올 때 다양한 방법을 통해 자료를 관찰하고 이해하는 과정

### 2-2.필요성
- 데이터의 분포 및 값을 검토함으로써 데이터가 표현하는 현상을 이해하며 내재된 잠재적 문제를 인식하고 해결안을 도출

### 2-3.분석 과정 및 절차

1.분석 목적과 변수 확인

2.데이터의 문제성 확인

3.데이터의 개별 속성값 분포 확인

4.데이터 사이의 관계 확인

### 2-4.데이터 문제성 확인

1.결측치 대치 방법

- 단순대치법, 다중대치법 등

2.이상치 대치 방법
- 제거, 대체, 유지 등

### 2-5.데이터의 개별 속성값 분포 확인

*적절한 요약 통계지표 사용*

1.데이터의 중심
- 평균(mean), 중앙값(median), 최빈값(mode)

2.데이터의 분산
- 범위(range), 분산(var), 표준편차(std)

*사분위범위 방법 사용 등*

## 3.파이썬 데이터 탐색

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 3-1.타이타닉 데이터셋

```python
df =pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
df
```

```python
df.info()
```

```python
#변수 타입변환(범주형 변수이므로 타입 변환)
df['Survived']=df['Survived'].astype(str)
df['Pclass']=df['Pclass'].astype(str)
```

```python
df.head()
```

```python
df.describe(include='all')
```

등급 변수 분석

```python
#그룹바이(), size()를 이용해 Pclass 빈도수 구하기
Pclass_grouped = df.groupby('Pclass')
Pclass_grouped.size()
```

요금 변수 분석

```python
#Histogram
plt.hist(df['Fare'])
plt.show
```

```python
#데이터 분리
data_0 = df[df['Survived']=='0']['Fare']
data_1 = df[df['Survived']=='1']['Fare']
```

```python
#Boxplot - Survived 값에 따른 Fare 값의 분포
fig, ax = plt.subplots()
ax.boxplot([data_0, data_1])
plt.show
```

성별 변수 분석

```python
#빈도 수 계산
Sex_grouped = df.groupby(df['Sex'])
print(Sex_grouped.size())
```

```python
#성별 생존자 수 - 여자
data_3 = df[df['Sex']=='female']['Survived']
Sex_1_grouped = pd.DataFrame(data_3).groupby(df['Survived'])
print(Sex_1_grouped.size())
print()

#성별 생존자 수 - 남자
data_4 = df[df['Sex']=='male']['Survived']
Sex_2_grouped = pd.DataFrame(data_4).groupby(df['Survived'])
print(Sex_2_grouped.size())
```
