## 1.넘파이

행렬이나 대규모 다차원 배열을 쉽게 처리하도록 도와주는 파이썬 패키지로 많은 머신러닝, 딥러닝 패키지가 넘파이를 기반으로 구현되어 있음

```python
import numpy as np #넘파이 패키지 추가
```

### 배열 생성 및 초기화


배열 생성

```python
ar1=np.array([1,2,3,4,5])
ar2=np.array([[1,2,3],[4,5,6]])


print(ar1)
print(ar2)
```

난수를 이용한 2차원 배열 생성

```python
ar3=np.random.randn(3)
ar4=np.random.randn(2,3)


print(ar3)
print(ar4)
```

원소를 0과 1로 초기화

```python
ar5=np.zeros(5)
ar6=np.ones((2,3))


print(ar5)
print(ar6)
```

배열 차수 변경

```python
ar7=np.arange(20,200,10)
ar8=ar7.reshape(3,6)

print(ar7)
print(ar8)
```

### 배열 다루기

```python
ar1=np.arange(1, 21, 1)
print(ar1)
print(ar1[1])

ar2=ar1.reshape(2,10)
print(ar2)

ar2[1][1]=100
print(ar2)
print(ar2[1][0], ar2[1][1], ar2[1][2])
```

배열 산술연산

배열의 연산은 배열 원소에 각각 적용

```python
ar1=np.arange(1, 11, 1)
ar2=ar1+3
ar3=ar1*2

print(ar1)
print(ar2, '\n', ar3)
```

배열 통계 메소드 사용


```python
ar1=np.array([[5,7,9], 
              [-7,-6,19],
              [6,9,11]])

print(ar1)
print(ar1.sum())
print(ar1.mean())
print(ar1.max())
print(ar1.min())
print(ar1.max(axis=0)) #세로로 비교해서 열의 원소 중 가장 큰 값을 구함
print(ar1.max(axis=1)) #가로로 비교해서 행의 원소 중 가장 큰 값을 구함
```

조건식을 이용한 배열 연산


```python
print(ar1>0)
m_count=(ar1<0).sum()
print(m_count)

ar2=np.where(ar1<0, 0, ar1) #ar1의 원소가 0보다 작으면 0으로 변경, 그렇지 않으면 ar1의 값으로 대입
print(ar2)
```

```python
ar1=np.array([[5,7,9], 
              [-3,-6,19],
              [6,4,11]])

print(ar1)
print()

ar1.sort(0) #열 단위로 원소 정렬
print(ar1)
print()

ar1.sort(1) #행 단위로 원소 정렬
print(ar1)
print()

ar1.sort() #default = 1
print(ar1)
```

*표준화*

각 데이터 값들이 평균을 기준으로 얼마나 떨어져 있는지 표현하는 값으로 변환하는 과정

Z-score은 각 데이터 값에서 평균을 밴 후 표준편차로 나누어 준 값(변환 후 평균은0, 표준편차는1)

numpy : z=(x-mean())/std()

scipy.stats : zscore()

sklearn.preprocessing : StandardScaler().fit_transform()

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/df_sample.csv')
df['합계']=df['중간']+df['기말']
df['cls']=x

mid_avg = np.mean(df['중간'])
mid_std = np.std(df['중간'])

df['중간 Z-score'] = (df['중간'] - mid_avg)/mid_std
df
```

*정규화*

데이터의 범위를 0과 1사이로 변환하여 데이터의 분포를 조정하는 것

(x-min())/(max()-min()) 또는 MinMaxScaler(), minmax_scale()함수를 이용

```python
import sklearn.preprocessing as sp
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/df_sample.csv')
df['합계']=df['중간']+df['기말']
df['cls']=x

mid_avg = np.mean(df['중간'])
mid_std = np.std(df['중간'])

df['중간 Z-score'] = (df['중간'] - mid_avg)/mid_std

fin_min = np.min(df['기말'])
fin_max = np.max(df['기말'])

df['기말_정규화'] = (df['기말']-fin_min)/(fin_max-fin_min)

df
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/df_sample.csv')
df['합계']=df['중간']+df['기말']
df['cls']=x

mid_avg = np.mean(df['중간'])
mid_std = np.std(df['중간'])

df['중간 Z-score'] = (df['중간'] - mid_avg)/mid_std

df['기말_정규화'] = sp.minmax_scale(df['기말'])
df
```

### 3-4.이상값 처리

이상값은 데이터의 범위에서 크게 벗어난 값

사분위수, 정규분포를 통해 식별하고 식별된 이상값은 제거 여부 판단

1QR-IQR보다 작거나 3QR+IQR보다 큰 수를 이상치로 판별

```python
df_1QR = df['기말'].quantile(0.25)
df_3QR = df['기말'].quantile(0.75)
df_IQR = df_3QR-df_1QR

print(df_1QR)
print(df_3QR)
print(df_IQR)
```

사분위수를 이용한 이상값 제거 

Q3 - Q1이 IQR이 되며, Q1에서 1.5 * IQR을 한 값을 빼준 값을 최소 제한선 

Q3에서 1.5 * IQR을 더해준 값을 최대 제한선으로 둔다.

```python
q_min = df_1QR - 1.5*df_IQR
q_max = df_3QR + 1.5*df_IQR

print(q_min, q_max)
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df['기말'])
plt.show()
```

```python
outlier_data = df[(df['기말'] < q_min) | (df['기말'] > q_max)].index
```

```python
df = df.drop(outlier_data, axis = 0)
df
```

```python
plt.boxplot(df['기말'])
plt.show()
```
