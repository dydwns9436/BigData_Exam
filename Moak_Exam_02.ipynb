## 제 2유형

### 데이터셋 : Parkinsons.csv

Q. 데이터셋을 이용해 파킨슨병을 예측하는 모델을 로지스틱 회귀모형을 적용해 생성하고, 이때 파킨슨병을 예측하는데 영향을 미치는 변수를 중요한 순서대로 3개 선정하시오.

> 분석조건

1. name 컬럼 제거 
2. 정규화는 최소최대정규화 적용
3. 로지스틱 회귀를 위해 상수항 추가
4. Status는 카테고리 타입으로 변환
5. 트레이닝셋과 테스트셋의 비율은 9:1
6. 최적화 방법론으로 'bfgs' 사용

```python
df = pd.read_csv('/content/drive/MyDrive/BigData/CSV_File/Parkinsons.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop('name',axis = 1)
```

```python
df.select_dtypes('object').columns
```

```python
df['status'] = df['status'].astype('category')
```

```python
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

mms = MinMaxScaler()

df_mms = mms.fit_transform(df)
df_mms = pd.DataFrame(df_mms)
```

```python
df_mms.columns = df.columns
```

```python
df_mms[['status']]
```

```python
# 상수항 추가 (상수항 추가는 왜하는지 더 배워봐야 겠다.)

import statsmodels.api as sm

df_mms = sm.add_constant(df_mms, has_constant = 'add')
```

```python
df_mms.tail()
```

```python
X = df_mms.drop('status', axis = 1)
y = df_mms['status']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify = y, random_state=20220621)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
lg = sm.Logit(y_train, X_train) #로지스틱 회귀 생성(여기서는 신기하게 종속변수가 먼저오고 독립변수가 그 후에 온다.)

sett = lg.fit(method = 'bfgs', maxiter = 1000)
```

```python
sett.summary()
```

```python
def cut_off(y, threshold):
    Y =y.copy()
    Y[Y>threshold]=1
    Y[Y<=threshold] = 0

    return(Y.astype(int))
```

```python
pred = sett.predict(X_test)
y_pred = cut_off(pred, 0.5)
```

```python
from sklearn.metrics import f1_score

f1_score(y_test, y_pred)
```
